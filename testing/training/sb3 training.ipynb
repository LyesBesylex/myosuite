```python
from stable_baselines3 import PPO
import os


logdir = "logs"
models_dir = "models/SACarmreach"

model = PPO("MlpPolicy", env, verbose=1,ent_coef=0.01,tensorboard_log = logdir)

if not os.path.exists(models_dir):
    os.makedirs(models_dir)

if not os.path.exists(logdir):
    os.makedirs(logdir)

print("========================================")
print("Starting policy learning")
print("========================================")

model.learn(total_timesteps=5000000)
 
print("========================================")
print("Job Finished.") 
print("========================================")
```

```python
import numpy as np
import myosuite
import gym
from stable_baselines3 import PPO
import time

env = gym.make('ArmReachFixed-v0')

model.load("./ArmReach_500000_PPO")

# evaluate policy

all_rewards = []
for _ in range(50): # 20 random targets
  ep_rewards = []
  done = False
  obs = env.reset()
  while not done:
      env.mj_render()
      # get the next action from the policy
      action, _ = model.predict(obs)
      # take an action based on the current observation
      obs, reward, done, info = env.step(action)
      ep_rewards.append(reward)
  all_rewards.append(np.sum(ep_rewards))
print(f"Average reward: {np.mean(all_rewards)} over 20 episodes")
```

```python
#Evaluate untrained policy

all_rewards = []
for _ in range(20): # 20 random targets
  ep_rewards = []
  done = False
  obs = env.reset()
  while not done:
      env.mj_render()
      # take a random action
      obs, reward, done, info = env.step(env.action_space.sample())
      ep_rewards.append(reward)
  all_rewards.append(np.sum(ep_rewards))
print(f"Average reward: {np.mean(all_rewards)} over 20 episodes")
```
